```python
from sklearn import preprocessing
lbl = preprocessing.LabelEncoder()
df['serialno'] = lbl.fit_transform(df['serialno'].astype(str))
```

```python

X = df.drop(columns=['Failure','serialno'])
y = df['Failure']
```

```python
import xgboost
import shap

# load JS visualization code to notebook
shap.initjs()

# train XGBoost model

model = xgboost.train({"learning_rate": 0.89}, xgboost.DMatrix(X, label=y), 400)

# explain the model's predictions using SHAP
# (same syntax works for LightGBM, CatBoost, scikit-learn and spark models)
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X)

# visualize the first prediction's explanation (use matplotlib=True to avoid Javascript)
shap.force_plot(explainer.expected_value, shap_values[0,:], X.iloc[0,:])
```

```python
# create a dependence plot to show the effect of a single feature across the whole dataset
shap.dependence_plot("reallocated_sectors_count_normalized", shap_values, X)
```

```python
shap.summary_plot(shap_values, X)
```

```python
shap.summary_plot(shap_values, X, plot_type="bar")
```

```python
shap_interaction_values = shap.TreeExplainer(model).shap_interaction_values(X.iloc[:400,:])

shap.summary_plot(shap_interaction_values, X.iloc[:400,:])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
df = pd.read_csv("hdd_datasetv2.csv")

nunique = df.apply(pd.Series.nunique)
cols_to_drop = nunique[nunique == 1].index
df.drop(cols_to_drop, axis=1, inplace=True)


f, ax = plt.subplots(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, ax=ax)
```
